```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
tweets_list_lrc = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterHashtagScraper('#loopring since:2022-09-21').get_items()):
    tweets_list_lrc.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
    
for i,tweet in enumerate(sntwitter.TwitterHashtagScraper('#lrc since:2022-09-21').get_items()):
    tweets_list_lrc.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
    
for i,tweet in enumerate(sntwitter.TwitterHashtagScraper('#LRC since:2022-09-21').get_items()):
    tweets_list_lrc.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
    
# Creating a dataframe from the tweets list above
tweets_df_lrc = pd.DataFrame(tweets_list_lrc, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])

tweets_df_lrc.size
tweets_df_lrc_stored = tweets_df_lrc 
```

```python
tweets_df_lrc.to_csv("loopring_tweets_september.csv", sep=',')
```

```python
tweets_df_lrc = tweets_df_lrc_stored
tweets_df_lrc.head()
tweets_df_lrc.set_index('Datetime')
tweets_df_lrc.index = pd.to_datetime(tweets_df_lrc['Datetime'])
print(tweets_df_lrc.size)
startDate = pd.to_datetime('2022-09-21 23:30:00+00:00')
tweets_df_lrc = tweets_df_lrc[tweets_df_lrc['Datetime'] >= startDate]
```

```python
from datetime import timedelta, datetime
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

months = {
    'January': 31,
    'February': 28,
    'March': 31,
    'April': 30,
    'May': 31,
    'June': 30,
    'July': 31,
    'August':31,
    'September':30,
    'October': 31,
    'November': 30,
    'December': 31,
}

# months = [31,28, 31,30,31,30,31,31,30,31, 30,31]

analyzed_tweets = pd.DataFrame(columns=['Timewindow', 'Sentiment', 'Volume'])

for year in range(2022, datetime.now().year + 1):
    tweets_year = tweets_df_lrc[tweets_df_lrc['Datetime'].dt.year == year]
    for month in months:
        tweets_month = tweets_year[pd.to_datetime(tweets_year['Datetime']).dt.strftime('%B') == month]
        for day in range(1, months[month] + 1):
            tweets_day = tweets_month.loc[(tweets_month['Datetime'].dt.day == day)]

            if (tweets_day.size == 0):
                continue
            
            for i in range(1, 49):
                start = str(timedelta(minutes=(i-1)*30))
                end = str(timedelta(minutes=(i*30)-1, seconds=59))
                current_tweets = tweets_day.between_time(start, end)
                results_tweet = []
                for tweet in current_tweets.iterrows():
                    sentiment_result = analyzer.polarity_scores(tweet[1]['Text'])
                    results_tweet.append(sentiment_result['compound'])
                
                sentiment = 0
                volume = 0
                timewindow = '{}-{}-{} {}-{}'.format(day, month, year, start, end)

                if (len(results_tweet) > 0):
                    sentiment = sum(results_tweet) / len(results_tweet)
                    volume = len(results_tweet)
                
                entry = {
                    'Timewindow': timewindow,
                    'Sentiment': sentiment,
                    'Volume': volume
                }
                analyzed_tweets = analyzed_tweets.append(entry, ignore_index=True)
        
print(analyzed_tweets.size / 3)
analyzed_tweets.to_csv('analzyed.csv', sep=',')
```
